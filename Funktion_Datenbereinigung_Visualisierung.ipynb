# Funktion f√ºr Datenbereinigung und Visualisierung

Erstellen Sie eine Funktion in Python, um einen Datensatz zu bereinigen: Gruppieren Sie den Datensatz nach Postleitzahl und berechnen Sie den Mittelwert f√ºr numerische Variablen und den Modus f√ºr kategoriale Variablen.

````python
‚úÖ Funktion: Datensatz bereinigen & aggregieren nach PLZ

```python
import pandas as pd
from scipy.stats import mode

def bereinige_und_agg_plz(df, plz_col='PLZ'):
    """
    Bereinigt den Datensatz:
    - Gruppiert nach Postleitzahl (PLZ)
    - Berechnet den Mittelwert f√ºr numerische Spalten
    - Berechnet den Modus (h√§ufigster Wert) f√ºr kategorische Spalten

    Gibt den aggregierten DataFrame zur√ºck.
    """
    # Numerische und kategorische Spalten erkennen
    num_cols = df.select_dtypes(include='number').columns.tolist()
    cat_cols = df.select_dtypes(include='object').columns.tolist()

    # Gruppierung nach PLZ
    grouped = df.groupby(plz_col)

    # Numerische Werte: Mittelwert berechnen
    df_num = grouped[num_cols].mean()

    # Kategorische Werte: Modus (h√§ufigster Wert)
    df_cat = grouped[cat_cols].agg(lambda x: mode(x, nan_policy='omit')[0][0])

    # Index zur√ºcksetzen und zusammenf√ºhren
    df_final = pd.concat([df_num, df_cat], axis=1).reset_index()

    return df_final
```
üì• Anwendung

```python
Beispiel-Datensatz laden
df = pd.read_csv("kunden.csv")

Funktion aufrufen
df_cleaned = bereinige_und_agg_plz(df)

Ausgabe anzeigen
print(df_cleaned.head())
```
````

Nimm einen Monat als Ausgangspunkt und stelle die Gesamtzahl der Nutzer als Balkendiagramm dar. Berechne f√ºr den n√§chsten Monat die Anzahl der neuen Nutzer, die Anzahl der abgewanderten Nutzer und stelle alles als Balkendiagramm dar, d.h. die Gesamtzahl des letzten Monats wird als Anzahl dargestellt und dann die Anzahl der neu hinzugekommenen Nutzer, die Anzahl der abgewanderten Nutzer und die Gesamtzahl.
````python
F√ºr die Berechnung ber√ºcksichtige die Kunden_ID
17.04.25, 20:33 - ChatGPT: Top Use Case f√ºr *Explorative Datenanalyse (EDA)* im Kundenverhalten! üîç
Du willst die *Nutzerinteraktionen im Monatsverlauf* analysieren ‚Äì mit Fokus auf:

- üìå *Gesamtanzahl Nutzer pro Monat*
- üü¢ *Neue Nutzer* (Zuwanderung)
- üî¥ *Abgewanderte Nutzer* (Auswanderung)

Das Ganze soll visuell als Barplot √ºber *12 Monate* dargestellt werden ‚Äì nice!

---

üß† Vorgehen

Wir brauchen:

1. Einen Datensatz mit einer *Spalte `Kunde_ID`* und einer *Datumsangabe*
2. Einen Startmonat
3. Eine *rollierende Berechnung* f√ºr:
   - Nutzer gesamt
   - Neue Nutzer
   - Nutzer, die den Dienst nicht mehr nutzen (nicht mehr aktiv im n√§chsten Monat)

---
````

````python
‚úÖ Python-Funktion zur Analyse & Visualisierung

```python
import pandas as pd
import matplotlib.pyplot as plt

def nutzerentwicklung_visualisieren(df, datum_col='Datum', id_col='Kunde_ID'):
    """
    Visualisiert die Nutzerentwicklung √ºber 12 Monate:
    - Gesamtanzahl der Nutzer
    - Neue Nutzer (Zuwanderung)
    - Abgewanderte Nutzer (Auswanderung)
    """
    df[datum_col] = pd.to_datetime(df[datum_col])
    df['Monat'] = df[datum_col].dt.to_period('M')

    monate = sorted(df['Monat'].unique())
    daten = []

    for i in range(len(monate) - 1):
        monat_aktuell = monate[i]
        monat_nachher = monate[i+1]
        # Aktuelle und n√§chste Nutzergruppe
        user_aktuell = set(df[df['Monat'] == monat_aktuell][id_col])
        user_nachher = set(df[df['Monat'] == monat_nachher][id_col])

        # Neue & abgewanderte Nutzer
        neue_nutzer = user_nachher - user_aktuell
        abgewandert = user_aktuell - user_nachher
        gesamt = len(user_aktuell)

        daten.append({
            'Monat': str(monat_aktuell),
            'Gesamt': gesamt,
            'Zuwanderung': len(neue_nutzer),
            'Abwanderung': len(abgewandert)
        })

    df_plot = pd.DataFrame(daten)

    # üìä Visualisierung
    plt.figure(figsize=(14, 6))
    plt.bar(df_plot['Monat'], df_plot['Gesamt'], label='Gesamt', color='gray')
    plt.bar(df_plot['Monat'], df_plot['Zuwanderung'], label='Zuwanderung', color='green', bottom=df_plot['Gesamt'])
    plt.bar(df_plot['Monat'], -df_plot['Abwanderung'], label='Abwanderung', color='red')

    plt.axhline(0, color='black', linewidth=0.8)
    plt.xticks(rotation=45)
    plt.ylabel('Anzahl Kunden')
    plt.title('Kundenentwicklung (12 Monate)')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
````

````python

üì• Anwendung

```python
df = pd.read_csv("kunden.csv")  # Enth√§lt Kunde_ID und Datum
nutzerentwicklung_visualisieren(df)
```
````

````python
üìä Interaktive Visualisierung mit Plotly

üîß Voraussetzungen

Stelle sicher, dass du die erforderlichen Bibliotheken installiert hast:


```bash
pip install pandas plotly
```


üß™ Funktion zur Analyse und Visualisierung

```python
import pandas as pd
import plotly.graph_objects as go

def nutzerentwicklung_visualisieren_interaktiv(df, datum_col='Datum', id_col='Kunde_ID'):
    """
    Visualisiert die Nutzerentwicklung √ºber 12 Monate interaktiv:
    - Gesamtanzahl der Nutzer
    - Neue Nutzer (Zuwanderung)
    - Abgewanderte Nutzer (Auswanderung)
    """
    df[datum_col] = pd.to_datetime(df[datum_col])
    df['Monat'] = df[datum_col].dt.to_period('M')

    monate = sorted(df['Monat'].unique())
    daten = []

    for i in range(len(monate) - 1):
        monat_aktuell = monate[i]
        monat_nachher = monate[i+1]
        # Aktuelle und n√§chste Nutzergruppe
        user_aktuell = set(df[df['Monat'] == monat_aktuell][id_col])
        user_nachher = set(df[df['Monat'] == monat_nachher][id_col])

        # Neue & abgewanderte Nutzer
        neue_nutzer = user_nachher - user_aktuell
        abgewandert = user_aktuell - user_nachher
        gesamt = len(user_aktuell)

        daten.append({
            'Monat': str(monat_aktuell),
            'Gesamt': gesamt,
            'Zuwanderung': len(neue_nutzer),
            'Abwanderung': len(abgewandert)
        })

    df_plot = pd.DataFrame(daten)

    # üìä Interaktive Visualisierung mit Plotly
    fig = go.Figure()
    fig.add_trace(go.Bar(
        x=df_plot['Monat'],
        y=df_plot['Gesamt'],
        name='Gesamt',
        marker_color='gray'
    ))
    fig.add_trace(go.Bar(
        x=df_plot['Monat'],
        y=df_plot['Zuwanderung'],
        name='Zuwanderung',
        marker_color='green'
    ))
    fig.add_trace(go.Bar(
        x=df_plot['Monat'],
        y=-df_plot['Abwanderung'],
        name='Abwanderung',
        marker_color='red'
    ))

    fig.update_layout(
        title='Kundenentwicklung (12 Monate)',
        xaxis_title='Monat',
        yaxis_title='Anzahl Kunden',
        barmode='relative',
        bargap=0.15,
        bargroupgap=0.1,
        template='plotly_white'
    )

    fig.show()
```
````

````python
Statt die Zuwanderung und Abwanderung *relativ zu Gesamt* (also gestapelt) darzustellen, kannst du *alle Werte unabh√§ngig als Balken zeigen*, damit man direkt sieht:

- üì¶ *Gesamtanzahl Nutzer im Monat*
- üü¢ *Neue Kunden (Zuwanderung)*
- üî¥ *Verlorene Kunden (Abwanderung)*

---

üîÑ Neue Version: *Absolute Fallzahlen als nebeneinanderliegende Balken*

Hier kommt die angepasste *Plotly-Visualisierung*:

```python
import pandas as pd
import plotly.graph_objects as go

def nutzerentwicklung_absolut(df, datum_col='Datum', id_col='Kunde_ID'):
    """
    Visualisiert absolute Nutzerzahlen je Monat:
    - Gesamtanzahl der Kunden
    - Neue Kunden (Zuwanderung)
    - Verlorene Kunden (Abwanderung)
    """
    df[datum_col] = pd.to_datetime(df[datum_col])
    df['Monat'] = df[datum_col].dt.to_period('M')

    monate = sorted(df['Monat'].unique())
    daten = []

    for i in range(len(monate) - 1):
        monat_aktuell = monate[i]
        monat_nachher = monate[i + 1]

        kunden_aktuell = set(df[df['Monat'] == monat_aktuell][id_col])
        kunden_nachher = set(df[df['Monat'] == monat_nachher][id_col])

        neue_kunden = kunden_nachher - kunden_aktuell
        abgewanderte_kunden = kunden_aktuell - kunden_nachher
        gesamt = len(kunden_aktuell)
        daten.append({
            'Monat': str(monat_aktuell),
            'Gesamt': gesamt,
            'Zuwanderung': len(neue_kunden),
            'Abwanderung': len(abgewanderte_kunden)
        })

    df_plot = pd.DataFrame(daten)

    # üìä Plotly: Nebeneinanderliegende Balken (barmode = group)
    fig = go.Figure(data=[
        go.Bar(name='Gesamt', x=df_plot['Monat'], y=df_plot['Gesamt'], marker_color='steelblue'),
        go.Bar(name='Zuwanderung', x=df_plot['Monat'], y=df_plot['Zuwanderung'], marker_color='green'),
        go.Bar(name='Abwanderung', x=df_plot['Monat'], y=df_plot['Abwanderung'], marker_color='crimson')
    ])

    fig.update_layout(
        title='Nutzerbewegung ‚Äì Absolute Zahlen je Monat',
        xaxis_title='Monat',
        yaxis_title='Anzahl Kunden',
        barmode='group',  # nebeneinander statt gestapelt
        bargap=0.2,
        template='plotly_white'
    )

    fig.show()
```
````

````python
Ich m√∂chte dazu auch gleich *Retention-Rate*, *Churn-Rate* oder *Kundenwachstum in Prozent* mit anzeigen? üí°

Dann erweitern wir die Funktion, um zus√§tzlich folgende *KPI-Metriken* pro Monat zu berechnen und darzustellen:

---

üìà Was wird neu berechnet?

| Metrik             | Bedeutung                                             |
|--------------------|--------------------------------------------------------|
| üìä *Gesamt*      | Anzahl Kunden im aktuellen Monat                      |
| üü¢ *Zuwanderung* | Neue Kunden, die im aktuellen Monat hinzugekommen sind |
| üî¥ *Abwanderung* | Kunden, die im n√§chsten Monat nicht mehr aktiv sind    |
| üîÅ *Retention-Rate* | Anteil der Kunden, die geblieben sind (im n√§chsten Monat wieder da) |
| ‚ùå *Churn-Rate*  | Anteil der Kunden, die abgewandert sind               |
| üìà *Wachstum (%)* | Netto-Ver√§nderung gegen√ºber dem Vormonat              |
````

````python

‚úÖ Funktion: Nutzerentwicklung + KPI-Analyse

```python
import pandas as pd
import plotly.graph_objects as go

def nutzerentwicklung_kpi(df, datum_col='Datum', id_col='Kunde_ID'):
    df[datum_col] = pd.to_datetime(df[datum_col])
    df['Monat'] = df[datum_col].dt.to_period('M')

    monate = sorted(df['Monat'].unique())
    daten = []

    for i in range(len(monate) - 1):
        monat_aktuell = monate[i]
        monat_nachher = monate[i + 1]
        kunden_aktuell = set(df[df['Monat'] == monat_aktuell][id_col])
        kunden_nachher = set(df[df['Monat'] == monat_nachher][id_col])

        neue_kunden = kunden_nachher - kunden_aktuell
        abgewanderte_kunden = kunden_aktuell - kunden_nachher
        geblieben = kunden_aktuell & kunden_nachher

        gesamt = len(kunden_aktuell)
        retention = len(geblieben) / gesamt if gesamt else 0
        churn = len(abgewanderte_kunden) / gesamt if gesamt else 0
        wachstum = (len(kunden_nachher) - gesamt) / gesamt if gesamt else 0

        daten.append({
            'Monat': str(monat_aktuell),
            'Gesamt': gesamt,
            'Zuwanderung': len(neue_kunden),
            'Abwanderung': len(abgewanderte_kunden),
            'RetentionRate': round(retention * 100, 2),
            'ChurnRate': round(churn * 100, 2),
            'Wachstum': round(wachstum * 100, 2)
        })

    df_plot = pd.DataFrame(daten)

    # üìä Balkendiagramm f√ºr absolute Nutzerzahlen
    fig = go.Figure()
    fig.add_trace(go.Bar(name='Gesamt', x=df_plot['Monat'], y=df_plot['Gesamt'], marker_color='steelblue'))
    fig.add_trace(go.Bar(name='Zuwanderung', x=df_plot['Monat'], y=df_plot['Zuwanderung'], marker_color='green'))
    fig.add_trace(go.Bar(name='Abwanderung', x=df_plot['Monat'], y=df_plot['Abwanderung'], marker_color='crimson'))

    # üîÅ Sekund√§rachse f√ºr KPIs: Retention, Churn, Wachstum
    fig.update_layout(
        title='Nutzerentwicklung und KPIs je Monat',
        xaxis_title='Monat',
        yaxis_title='Anzahl Kunden',
        barmode='group',
        template='plotly_white'
    )

    # üìà KPI-Linienplot
    fig.add_trace(go.Scatter(
        name='RetentionRate (%)',
        x=df_plot['Monat'],
        y=df_plot['RetentionRate'],
        yaxis='y2',
        mode='lines+markers',
        line=dict(color='orange', width=2, dash='dash')
    ))

    fig.add_trace(go.Scatter(
        name='ChurnRate (%)',
        x=df_plot['Monat'],
        y=df_plot['ChurnRate'],
        yaxis='y2',
        mode='lines+markers',
        line=dict(color='red', width=2)
    ))

    fig.add_trace(go.Scatter(
        name='Wachstum (%)',
        x=df_plot['Monat'],
        y=df_plot['Wachstum'],
        yaxis='y2',
        mode='lines+markers',
        line=dict(color='green', width=2)
    ))

    # Sekund√§rachse konfigurieren
    fig.update_layout(
        yaxis2=dict(
            title='Prozentwerte',
            overlaying='y',
            side='right'
        )
    )

    fig.show()
```
````

````python
üì• Anwendung

```python
df = pd.read_csv("kunden.csv")
nutzerentwicklung_kpi(df)
```
````
